```python
import pandas as pd

filename="./hospital_info.csv"
df = pd.read_csv(filename, encoding='utf-8')
```

```python
df.head()
```

```python
import re

json_data = []

for index, row in df.iterrows():
    address = row[2].split(',')
    address = re.sub(r'\(.*?\)', '', address[0])
    json_data.append(
        {
            '의료기관종별': row[0],
            '의료기관명':row[1],
            '의료기관주소':address,
            '전화번호':row[3],
        }
    )
```

```python
import json

json_file_name = "hospital_v2.json"
with open(json_file_name, "w", encoding='utf-8') as json_file:
    json.dump(json_data, json_file, ensure_ascii=False, indent=4)
```

```python
import json
with open(json_file_name, "r", encoding="utf-8") as st_json:
    st_python = json.load(st_json)
```
